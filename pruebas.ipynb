```python
mz_array_list = loadmat("addon/mz_array_list.mat")['data']
```

```python
precision = 0.0001
cleaned_mz_array_list = []
counter = 0
for mz_array in mz_array_list:
    cleaned_mz_array = []
    previous_mz = None
    for mz in mz_array:
        if previous_mz is None or (previous_mz is not None and (abs(mz - previous_mz) > precision)):
            cleaned_mz_array.append(mz)
            previous_mz = mz
        else:
            counter += 1
    cleaned_mz_array_list.append(np.array(cleaned_mz_array))
arr = np.array(cleaned_mz_array_list).flatten()
arr2 = mz_array_list.flatten()
```
